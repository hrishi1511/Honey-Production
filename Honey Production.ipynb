```python
#Step 1
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\joshi\Downloads\honeyproduction (1).csv")
```

```python
#Step 3
production_per_year = df[['prodvalue','year']].groupby('year')['prodvalue'].sum()
```

```python
plt.figure(figsize=(8,8))
plt.pie(production_per_year,labels=production_per_year.index,autopct='%1.1f%%',startangle=90)
plt.title('Percentage Distribution of Production Value by Year')
plt.show()
```

```python
#Step 4
sns.displot(df['priceperlb'], kde=True)
plt.title('Distribution of price per lb')
plt.show()
```

```python
#Step 5
sns.scatterplot(x='numcol', y='prodvalue',data=df)
plt.title('No. of Honey Producing Colonies VS Production Value')
plt.show()
```

```python
#Step 6
plt.figure(figsize=(10,4))
sns.boxplot(y='prodvalue',x='year',data=df)
plt.title("Production Value w.r.t. Year")
plt.show()
```

```python
#Step 7
df.columns
```

```python
plt.figure(figsize=(8,3))
sns.pairplot(df[['numcol','yieldpercol','totalprod','prodvalue','year']],hue='year',palette='Set1')
plt.suptitle('Pair plot between multiple variables w.r.t. year',y=1.02)
plt.show()
```

```python
#Step 8
sns.heatmap(df[['numcol', 'yieldpercol','totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr())
plt.title("Correlation Matrix")
plt.show()
```
